```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set computational parameters
L = 40  # length of the interval we focus on: [-L/2, L/2]
nmesh = 400  # number of mesh points
a = L / nmesh  # spatial step size

# Set physical parameters
h = 1  # Planck's constant (hbar) (we set hbar = 1)
m = 1  # mass of the particle (we set m = 1)
gamma = h**2 / (2 * m * a**2)  # hopping rate (related to the potential)
```

```python
# Hamiltonian definition
# Step 1. We initialize the Hamiltonian matrix to zero
H = np.zeros((nmesh, nmesh))

# Step 2. We fill the upper diagonal
for n in range(nmesh - 1):
    H[n, n+1] = -gamma
    H[n+1, n] = -gamma
H[0, nmesh-1] = -gamma
H[nmesh-1, 0] = -gamma
# Step 3: fill the main diagonal
np.fill_diagonal(H, 2*gamma)

# Compute the eigenvalues of H
en = np.linalg.eigvalsh(H)
print(en)

# Plot the eigenenergies
plt.scatter(np.arange(len(en)),en)
plt.title("Eigenenergies of the one-dimensional harmonic oscillator")
plt.xlabel("Eigenstate index")
plt.ylabel("Energy")
plt.xlim(0, L)
plt.ylim(0, 4)
plt.grid(True)
plt.show()
```
